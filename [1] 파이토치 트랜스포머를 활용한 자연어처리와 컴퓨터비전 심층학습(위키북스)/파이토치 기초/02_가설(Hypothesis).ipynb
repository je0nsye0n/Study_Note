### [02. 가설]

- 가설(Hypothesis) : 어떤 사실을 설명하거나 증명하기 위한 가정으로 두 개 이상의 변수의 관계를 검증 가능한 형태로 기술하여 변수 간의 관계를 예측하는 것

- 머신러닝에서의 가설   
머신러닝에서의 가설은 통계적 가설 검정이 되며, 데이터와 변수 간의 관계가 있는지 확률론적으로 설명하게 된다. 즉, 머신러닝에서의 가설은 독립 변수(X)와 종속 변수(Y)를 가장 잘 매핑시킬 수 있는 기능을 학습하기 위해 사용된다.   
    
독립변수(X)와 종속변수(Y)가 주어진다면 주어진 데이터 (X,Y)를 기반으로 가설을 세울 수 있다. 아래 그림의 붉은색 점은 데이터를 나타내며 파란색 선은 가설을 의미한다. 이 파란색 선은 선형 회귀를 통해 계산된 값이며, 수식으로 나타내면 y=ax+b로 표현할 수 있다.   
이 때, 머신러닝에서는 H(x)=Wx+b로 나타내며 H(x),W,b를 각각 가설, 가중치, 편향을 의미한다.   
마지막으로 학습이 된 결과를 모델이라 부르며, 이 모델을 통해 새로운 입력에 대한 결괏값을 예측한다.   
    

###### 출처 : 나무위키   

```python
# 성별에 따른 키 차이 검정(3.9)

import numpy as np
import pandas as pd
import seaborn as sns
from scipy import stats
from matplotlib import pyplot as plt

man_height = stats.norm.rvs(loc=170, scale=10, size=500, random_state=1)
woman_height = stats.norm.rvs(loc=150, scale=10, size=500, random_state=1)

X=np.concatenate([man_height,woman_height])
Y=["man"]*len(man_height) + ["woman"]*len(woman_height)

df = pd.DataFrame(list(zip(X,Y)), columns=["X","Y"])
fig = sns.displot(data=df, x="X", hue="Y", kind="kde")
plt.show()
```
